```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os
```

```python
A = imread("datasets/dog.jpg")
```

```python
A.shape
```

```python
plt.imshow(A);
```

```python
X = np.mean(A, -1)  # conver to gray
img = plt.imshow(256-X)
matplotlib.image.imsave('dog_gray.jpg', X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X)
S = np.diag(S)
```

```python
S.shape
```

```python
j = 0
for r in (5, 20, 100):
    X_approx = U[:,:r]@S[:r, :r]@VT[:r,:]
    matplotlib.image.imsave(str(j)+'name.jpg', X_approx)
    plt.figure(j+1)
    j += 1
    img = plt.imshow(256 - X_approx)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()
```

```python
plt.figure(2)
plt.plot(np.cumsum(np.diag(S)/np.sum(np.diag(S))))
plt.title('Singular Values : Cumulative Sum')
plt.show()
```
