# 030.Decision Tree

### iris dataset

iris.feature_names : 

Sepal Length : 꽃받침 길이  
Sepal Width  : 꽃받침 너비   
Petal Length  : 꽃잎 길이  
Petal Width   :  꽃잎 너비 

Species (꽃의 종류) :  setosa / versicolor / virginica 의 3종류로 구분된다.

**위 feature 를 모두 가지고 Decision Tree 알고리즘을 이용하여 꽃의 종류 분류**

```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
```

```python
iris.data.shape
```

```python
iris.feature_names
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2)
```

```python
clf.fit(iris.data, iris.target)
```

### graphviz 설치

- conda install python-graphviz  
- pip install pydotplus

```python
import graphviz 

dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names,  class_names=iris.target_names,  
                                filled=True, rounded=True, special_characters=True)  
```

### gini - 지니 불순도

지니 불순도는 집합에 이질적인 것이 얼마나 섞였는지를 측정하는 지표이며 CART (Classification And Regression Tree) 알고리즘에서 사용한다. 

집합에 있는 항목이 모두 같다면 지니 불순도는 최솟값(0)을 갖게 되며 이 집합은 완전히 순수하다고 할 수 있다.

```python
graphviz.Source(dot_data) 
```

```python
clf = tree.DecisionTreeClassifier(max_depth=None)
clf.fit(iris.data, iris.target)

dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names,  class_names=iris.target_names,  
                                filled=True, rounded=True, special_characters=True)  
graphviz.Source(dot_data) 
```

### train / test split & accuracy check

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
clf = tree.DecisionTreeClassifier(max_depth=None)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
