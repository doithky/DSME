# 085.PCA (Principal Component Analysis)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### 통신회사 고객 이탈 (Churn) data 를 이용

- 각 행은 고객을 나타내고 각 열은 고객 속성을 포함 :  전화, 여러 회선, 인터넷, 온라인 보안, 온라인 백업, 장치 보호, 기술 지원, TV 및 영화 스트리밍 등 각 고객이 가입 한 서비스, 계약, 지불 방법, 종이없는 청구, 월별 청구 및 총 청구 기간, 성별, 연령대 및 파트너 및 부양 가족이있는 경우에 관한 인구 통계 정보


- Churn : 지난 달에 떠난 고객 - 1

```python
churn_df = pd.read_csv("datasets/ChurnData.csv")
```

```python
churn_df.head()
```

```python
churn_df.cov()
```

```python
A = np.array(churn_df.values)
A.shape
```

```python
columns = churn_df.columns[:-1]
columns.size
```

```python
X = np.asarray(churn_df.loc[:, columns], dtype=np.float32)
y = np.asarray(churn_df['churn'])
```

```python
# train / test dataset split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### PCA 적용 전 Logistic Regression

- PCA 적용 후의 결과와 비교 목적

```python
# Fitting Logistic Regression
clf = LogisticRegression(solver='lbfgs', random_state=0)
clf.fit(X_train, y_train)
```

```python
# predict test set
y_pred = clf.predict(X_test)
# Confusion matrix
accuracy_score(y_test, y_pred)
```

## PCA 적용

- 특이값 분해 (Singular Value Decomposition, svd) 를 이용하여 27 개의 feature 를 2 개로 차원 축소  


#### Attributes

- components_  
    - shape : (n_components, n_features)
    - n_feature dimension 공간에서의 주성분 축 --> data 의 분산을 최대로 보존하는 방향

- explained_variance_ 
    - shape : (n_components,) 
    - 각 component 에 의해 설명된 variance 량
    
- explained_variance_ratio_ 
    - shape (n_components,)
    - 각 component 에 의해 설명된 variance 비율 


```python
# Apply kernel PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)                # 2 개 component 로 차원 축소

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print("차원 축소된 X_train :", X_train_pca.shape)
print(X_train_pca[:5])
print()
print("차원 축소된 X_test :", X_test_pca.shape)
print(X_test_pca[:5])
```

### Singular Value Decomposition 을 이용한 주성분 추출과 비교

```python
u, s, vh = np.linalg.svd(X_train)
print(u.shape)
print(s.shape)
print(vh.shape)
print(vh[0])
```

```python
print(pca.components_.shape)
print("첫번째 주성분 :")
print(pca.components_[0])
print("두번째 주성분 :")
print(pca.components_[1])
print('설명된 분산의 비율', pca.explained_variance_ratio_)
```

- 차원 축소된 data 를 이용한 logistic regression

```python
# Fitting Logistic Regression
clf = LogisticRegression(solver='lbfgs', random_state=0)
clf.fit(X_train_pca, y_train)
```

```python
# predict test set
y_pred = clf.predict(X_test_pca)
y_pred
```

```python
# Confusion matrix
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## 차원 축소된 churn data 시각화
- 27 개의 feature 가 2 개의 PCA 로 차원 축소 되었으므로 평면상의 시각화 가능

```python
X_train_pca.shape
```

- 차원축소된 X_train_pca 를 X1, X2 축으로 시각화  

- y_train = 0, 1 을 color 로 구분

```python
X_train_pca[y_train == 0][:5]
```

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['r', 'b'])

X1 = X_train_pca[y_train == 0, 0]
X2 = X_train_pca[y_train == 0, 1]
plt.scatter(X1, X2, color=cmap(0), label=0)

X1 = X_train_pca[y_train == 1, 0]
X2 = X_train_pca[y_train == 1, 1]
plt.scatter(X1, X2, color=cmap(1), label=1)

plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.title('Dimension Reduced churn data');
plt.show()
```
